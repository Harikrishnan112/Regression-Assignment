```
# SIMPLE LINEAR REGRESSION : LOAD vs SPRING EXTENSION

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# -----------------------------
# Experimental Values
# -----------------------------
# Load (N)
X = np.array([10, 20, 30, 40, 50, 60]).reshape(-1, 1)

# Extension (cm)
y = np.array([2.1, 4.0, 6.2, 8.1, 10.3, 12.2])

# -----------------------------
# Create Linear Regression Model
# -----------------------------
model = LinearRegression()

# Train the model
model.fit(X, y)

# -----------------------------
# Regression Parameters
# -----------------------------
slope = model.coef_[0]
intercept = model.intercept_

print("Slope (m) =", slope)
print("Intercept (c) =", intercept)

# Regression Equation
print(f"\nRegression Equation:")
print(f"Extension = {slope:.4f} × Load + {intercept:.4f}")

# -----------------------------
# Prediction for 55 N
# -----------------------------
load_value = np.array([[55]])

predicted_extension = model.predict(load_value)

print(f"\nPredicted Extension for 55 N load = {predicted_extension[0]:.4f} cm")

# -----------------------------
# Plotting
# -----------------------------
plt.scatter(X, y, color='blue', label='Experimental Data')

# Regression Line
plt.plot(X, model.predict(X), color='red', linewidth=2,
         label='Regression Line')

# Labels and Title
plt.xlabel("Load (N)")
plt.ylabel("Extension (cm)")
plt.title("Simple Linear Regression: Load vs Spring Extension")

plt.legend()
plt.grid(True)

# Show graph
plt.show()
```
